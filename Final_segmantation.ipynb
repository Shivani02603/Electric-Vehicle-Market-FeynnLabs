```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Visualization style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

print("\nFINAL MARKET SEGMENTATION ANALYSIS - INTEGRATED MODEL\n" + "="*80)

# =========================
# MEMBER A: CAR ANALYSIS
# =========================
car_segments = pd.DataFrame({
    'Vehicle_Type': ['Hatchback', 'Sedan', 'SUV'],
    'Market_Share_%': [42.80, 26.57, 24.23],
    'Avg_Price_Lakhs': [3.80, 8.16, 13.46],
    'Overall_Score': [99.95, 62.71, 59.56]
})

# =========================
# MEMBER B: BIKE ANALYSIS
# =========================
bike_segments = pd.DataFrame({
    'Vehicle_Type': ['Cruiser Bike', 'Scooter', 'Commuter Bike'],
    'Market_Share_%': [22.65, 13.07, 10.28],
    'Avg_Price_Lakhs': [10.17, 1.76, 0.98],
    'Overall_Score': [95.16, 67.14, 57.99]
})

# =========================
# MEMBER C & D: CUSTOMER SEGMENTATION
# =========================
customer_segments = pd.DataFrame({
    'Cluster': [0, 1, 2, 3],
    'Segment_Label': ['D', 'B', 'C', 'A'],
    'Avg_Age': [31.7, 48.7, 37.9, 75.4],
    'Spending_Score_Value': [1.00, 1.93, 1.46, 2.07],
    'Work_Experience': [2.75, 2.24, 5.69, 1.19],
    'Family_Size': [3.03, 2.88, 2.33, 1.99],
    'Profession': ['Healthcare', 'Artist', 'Homemaker', 'Lawyer'],
    'Male_%': [49.7, 62.0, 17.5, 50.9],
    'Graduated_%': [50.9, 72.7, 56.9, 63.3]
}).sort_values('Avg_Age')

# =========================
# FINAL INTEGRATED MARKET SEGMENTS
# =========================
market_segments = pd.DataFrame({
    'Segment_Name': [
        'Young Family Commuters', 'Budget Urban Riders',
        'Mid-Career Professionals', 'Premium Bike Enthusiasts',
        'Retired Affluent (Limited)', 'Entry-Level Youth'
    ],
    'Primary_Vehicle': [
        'Electric Hatchback', 'Electric Scooter',
        'Electric Hatchback', 'Electric Cruiser Bike',
        'Not Targeted', 'Electric Commuter Bike (Phase 3)'
    ],
    'Customer_Cluster': ['C', 'C', 'B', 'B', 'A', 'D'],
    'Age_Group': ['30-40', '30-40', '45-55', '45-55', '70+', '25-35'],
    'Vehicle_Price': ['8-12L', '2-3L', '10-14L', '10-15L', '-', '1.5-2L'],
    'Launch_Priority': ['Phase 1 - HIGH', 'Phase 1 - MEDIUM', 'Phase 2',
                        'Phase 1 - HIGH', 'Not Targeted', 'Phase 3'],
    'Market_Size_%': [35, 15, 20, 20, 5, 5]
})

# =========================
# VISUALIZATIONS
# =========================
fig = plt.figure(figsize=(20, 14))

# 1. Market Share - All Vehicles
ax1 = plt.subplot(3, 4, 1)
all_vehicles = pd.concat([
    car_segments[['Vehicle_Type', 'Market_Share_%']],
    bike_segments[['Vehicle_Type', 'Market_Share_%']]
])
colors_v = ['#3498db', '#5dade2', '#85c1e9', '#e74c3c', '#ec7063', '#f1948a']
ax1.barh(all_vehicles['Vehicle_Type'], all_vehicles['Market_Share_%'],
         color=colors_v, edgecolor='black')
ax1.set_title('Market Share - All Vehicles', fontweight='bold')
ax1.set_xlabel('Market Share (%)')
ax1.invert_yaxis()

# 2. Price Positioning
ax2 = plt.subplot(3, 4, 2)
all_prices = pd.concat([
    car_segments[['Vehicle_Type', 'Avg_Price_Lakhs']],
    bike_segments[['Vehicle_Type', 'Avg_Price_Lakhs']]
])
ax2.bar(range(len(all_prices)), all_prices['Avg_Price_Lakhs'],
        color=colors_v, edgecolor='black')
ax2.set_xticks(range(len(all_prices)))
ax2.set_xticklabels(all_prices['Vehicle_Type'], rotation=45, ha='right')
ax2.set_title('Price Positioning', fontweight='bold')
ax2.set_ylabel('Price ( Lakhs)')

# 3. Customer Age Distribution
ax3 = plt.subplot(3, 4, 3)
colors_seg = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']
ax3.bar(customer_segments['Segment_Label'], customer_segments['Avg_Age'],
        color=colors_seg, edgecolor='black')
ax3.set_title('Average Age by Segment', fontweight='bold')
ax3.set_xlabel('Segment')
ax3.set_ylabel('Age')

# 4. Spending Power
ax4 = plt.subplot(3, 4, 4)
ax4.bar(customer_segments['Segment_Label'],
        customer_segments['Spending_Score_Value'],
        color=colors_seg, edgecolor='black')
ax4.set_title('Spending Power', fontweight='bold')
ax4.set_xlabel('Segment')
ax4.set_ylabel('Spending Score (1=Low, 3=High)')
ax4.axhline(y=2, color='red', linestyle='--', alpha=0.5)

# 5. Launch Priorities (Phase 1)
ax5 = plt.subplot(3, 4, 5)
phase1 = market_segments[market_segments['Launch_Priority'].str.contains('Phase 1')]
colors_p = ['#e74c3c' if 'HIGH' in p else '#3498db' for p in phase1['Launch_Priority']]
ax5.barh(phase1['Segment_Name'], phase1['Market_Size_%'],
         color=colors_p, edgecolor='black')
ax5.set_title('Phase 1 Launch Priorities', fontweight='bold')
ax5.set_xlabel('Market Size (%)')
ax5.invert_yaxis()

# 6. Market Segment Distribution
ax6 = plt.subplot(3, 4, 6)
pie_colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#95a5a6', '#34495e']
ax6.pie(market_segments['Market_Size_%'], labels=market_segments['Segment_Name'],
        autopct='%1.1f%%', colors=pie_colors, startangle=90, textprops={'fontsize':8})
ax6.set_title('Market Segment Distribution', fontweight='bold')

# 7. Vehicle-Customer Mapping
ax7 = plt.subplot(3, 4, 7)
mapping_data = {'Hatchback + C': 35, 'Scooter + C': 15,
                'Hatchback + B': 20, 'Cruiser + B': 20, 'Future': 10}
ax7.barh(list(mapping_data.keys()), list(mapping_data.values()),
         color=['#3498db', '#5dade2', '#85c1e9', '#e74c3c', '#95a5a6'],
         edgecolor='black')
ax7.set_title('Vehicle-Customer Mapping', fontweight='bold')
ax7.set_xlabel('Priority (%)')
ax7.invert_yaxis()

# 811. Work Experience, Family, Education, Gender
for i, (title, col, ylbl) in enumerate([
    ('Work Experience', 'Work_Experience', 'Years'),
    ('Family Size', 'Family_Size', 'Avg Size'),
    ('Education (% Graduated)', 'Graduated_%', '%'),
    ('Gender Distribution', 'Male_%', '% Male')
], start=8):
    ax = plt.subplot(3, 4, i)
    ax.bar(customer_segments['Segment_Label'], customer_segments[col],
           color=colors_seg, edgecolor='black')
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Segment')
    ax.set_ylabel(ylbl)

# 12. Launch Strategy (Text)
ax12 = plt.subplot(3, 4, 12)
ax12.axis('off')
ax12.text(0, 1, """LAUNCH STRATEGY TIMELINE

Phase 1: Hatchback (35%), Cruiser (20%), Scooter (15%)
Phase 2: Hatchback (20%)
Phase 3: Commuter Bike (5%)
Not Targeted: Cluster A (Retired)""",
          transform=ax12.transAxes, fontsize=9, va='top',
          bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

plt.suptitle('EV Market Segmentation - Complete Integration', fontsize=17, fontweight='bold')
plt.tight_layout()
plt.savefig('final_integrated_segmentation.png', dpi=300, bbox_inches='tight')
plt.show()

# =========================
# STRATEGIC RECOMMENDATIONS
# =========================
print("\nSTRATEGIC RECOMMENDATIONS\n" + "="*80)
print("""
PHASE 1 (Year 1):
- Electric Hatchback  Cluster C (35%)
- Electric Cruiser Bike  Cluster B (20%)
- Electric Scooter  Cluster C (15%)

PHASE 2 (Year 2-3):
- Electric Hatchback  Cluster B (20%)

PHASE 3 (Year 3+):
- Electric Commuter Bike  Cluster D (5%)

NOT TARGETED:
- Cluster A (Retired, Age 75+)

INSIGHT:
Cluster C (Age 3040, ~6 years work exp) is the SWEET SPOT:
aligns with top vehicles (Hatchback + Scooter) = 55%+ market access.
""")
print("="*80)
print("Visualization saved as: final_integrated_segmentation.png")
```
